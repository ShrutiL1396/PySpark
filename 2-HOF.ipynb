```python
import pandas as pd
from functools import reduce
import math
import numpy as np
```

```python
df = pd.read_csv("citibike.csv")
rows = df.values.tolist()
```

```python
rows
```

## 2a)

```python
# extracting the column containing 'gender' from the csv file.
gender = list(map(lambda x:x[16],rows)) 

# checking whether gender is '1' or '2' and storing the result in a list.
pairs = list(map(lambda x: (int(x==1),int(x==2)),gender))
```

```python
# uding reduce function to find the total trips for '1' gender and '2' gender
trip_no = reduce(lambda x,y: (x[0] + y[0], x[1] + y[1]),pairs)
print("The number of trips for gender(1) and gender(2) are:- {} respectively".format(trip_no))
```
